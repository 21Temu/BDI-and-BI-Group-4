```python
"""
PHASE 2: MAPREDUCE LOGIC (Python Implementation)
Jobs: Average Rating, User Review Count, Top 10 Products
"""

import gzip
import json
from collections import defaultdict
import heapq

print("="*60)
print("PHASE 2 - MAPREDUCE RESULTS")
print("="*60)

# Load data
reviews = []
with gzip.open('Electronics_sample_50k.json.gz', 'rt', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i >= 50000:
            break
        reviews.append(json.loads(line))

print(f"✅ Loaded {len(reviews):,} reviews")

# ============================================
# JOB 1: Average Rating Per Product (like MapReduce)
# ============================================

print("\n" + "="*60)
print("JOB 1: Average Rating Per Product")
print("="*60)

# MAP phase: extract (product, rating)
product_ratings = defaultdict(list)
for review in reviews:
    product = review.get('asin', 'unknown')
    rating = review.get('overall', 0)
    product_ratings[product].append(rating)

# REDUCE phase: calculate average
product_avg = {}
for product, ratings in product_ratings.items():
    product_avg[product] = sum(ratings) / len(ratings)

# Show top 10
top_products = sorted(product_avg.items(), key=lambda x: x[1], reverse=True)[:10]
print("\nTop 10 Products by Average Rating:")
for i, (product, avg) in enumerate(top_products, 1):
    print(f"   {i}. {product}: {avg:.2f} stars")

# ============================================
# JOB 2: Reviews Per User (Filtering & Grouping)
# ============================================

print("\n" + "="*60)
print("JOB 2: Review Count Per User")
print("="*60)

# MAP phase: extract (user, 1)
user_counts = defaultdict(int)
for review in reviews:
    user = review.get('reviewerID', 'unknown')
    user_counts[user] += 1

# REDUCE phase: sum counts
top_users = sorted(user_counts.items(), key=lambda x: x[1], reverse=True)[:10]
print("\nTop 10 Most Active Reviewers:")
for i, (user, count) in enumerate(top_users, 1):
    print(f"   {i}. {user}: {count} reviews")

# ============================================
# JOB 3: Top 10 Most Reviewed Products (Advanced)
# ============================================

print("\n" + "="*60)
print("JOB 3: Top 10 Most Reviewed Products")
print("="*60)

# MAP phase: extract (product, 1)
product_counts = defaultdict(int)
for review in reviews:
    product = review.get('asin', 'unknown')
    product_counts[product] += 1

# REDUCE phase: find top 10
top_10 = heapq.nlargest(10, product_counts.items(), key=lambda x: x[1])
print("\nTop 10 Most Reviewed Products:")
for i, (product, count) in enumerate(top_10, 1):
    avg_rating = product_avg.get(product, 0)
    print(f"   {i}. {product}: {count} reviews (avg: {avg_rating:.2f})")

# ============================================
# Save Results
# ============================================

with open('phase2_results.txt', 'w') as f:
    f.write("PHASE 2 - MAPREDUCE RESULTS\n")
    f.write("="*60 + "\n\n")
    f.write("JOB 1: Average Rating Per Product (Top 10)\n")
    for i, (product, avg) in enumerate(top_products, 1):
        f.write(f"{i}. {product}: {avg:.2f} stars\n")
    f.write("\nJOB 2: Review Count Per User (Top 10)\n")
    for i, (user, count) in enumerate(top_users, 1):
        f.write(f"{i}. {user}: {count} reviews\n")
    f.write("\nJOB 3: Top 10 Most Reviewed Products\n")
    for i, (product, count) in enumerate(top_10, 1):
        avg_rating = product_avg.get(product, 0)
        f.write(f"{i}. {product}: {count} reviews (avg: {avg_rating:.2f})\n")

print("\n✅ Results saved to 'phase2_results.txt'")
print("\n✅ PHASE 2 COMPLETE! Take screenshots of the outputs above.")
```
